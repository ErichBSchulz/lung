```python
from matplotlib import pyplot as plt
import numpy as np, math
from lung import volume_from_pressure_vectorized as volume_from_pressure, pressure_from_volume_vectorized as pressure_from_volume
```

```python
%matplotlib inline
```

```python
v=np.arange(0, 100, 1)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
plt.ylim(ymax = 40, ymin=-40)
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
ax.plot(v, pressure_from_volume(v, "Lung"), "o", label = "Lung")
ax.plot(v, pressure_from_volume(v, "Chest"), "v", label = "Chest")
ax.plot(v, pressure_from_volume(v, "Total"), "^", label = "Total")
plt.legend(loc='upper left')
ax.set_title("Compliance")
ax.set_xlabel("% of VC")
ax.set_ylabel("Pressure (cm of H2O)")
```

```python
p=np.arange(-40, 40, 0.05)
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
plt.ylim(ymax = 100, ymin=0)
plt.grid(b=True, which='major', color='gray', alpha=0.6, linestyle='dashdot', lw=1.5)
ax.plot(p, volume_from_pressure(p, "Lung"), "o", label = "Lung")
ax.plot(p, volume_from_pressure(p, "Chest"), "v", label = "Chest")
ax.plot(p, volume_from_pressure(p, "Total"), "^", label = "Total")
plt.legend(loc='upper left')
ax.set_title("Compliance")
ax.set_ylabel("% of VC")
ax.set_xlabel("Pressure (cm of H2O)")
```
